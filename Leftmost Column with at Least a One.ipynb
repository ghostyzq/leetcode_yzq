```python

class Solution:
    def leftMostColumnWithOne(self, binaryMatrix: 'BinaryMatrix') -> int:
        index, min_column = 10000000,1000000
        [n,m] = binaryMatrix.dimensions()
        print(n)
        for i in range(m):
            count = 0
            for j in range(n):
                tmp = binaryMatrix.get(j,i)
                if tmp == 1:
                    count += 1
            if count < min_column and count != 0:
                index = i
                min_column = count
        if index == 10000000:
            return -1
        return index
```

```python

class Solution:
    def leftMostColumnWithOne(self, bm: 'BinaryMatrix') -> int:
            n,m = bm.dimensions();
            i = n-1
            j= m-1
            while (i>=0 and j>=0):
                if bm.get(i,j)==0:
                    i-=1
                else:
                    j-=1
            if j == m-1:
                return -1
            return j+1
```
