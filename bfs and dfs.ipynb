```python
class Solution:
    def bfs(self, graph, s):
        queue = []
        queue.append(s)
        visited = set()
        visited.add(s)
        while queue:
            vertex = queue.pop(0)
            nodes = graph[vertex]
            for x in nodes:
                if x not in visited:
                    queue.append(x)
                    visited.add(x)
            print(vertex)
    def dfs(self, graph, s):
        stack = []
        stack.append(s)
        visited = set()
        visited.add(s)
        while stack:
            vertex = stack.pop(len(stack)-1)
            nodes = graph[vertex]
            for x in nodes:
                if x not in visited:
                    stack.append(x)
                    visited.add(x)
            print(vertex)
                
        
```
