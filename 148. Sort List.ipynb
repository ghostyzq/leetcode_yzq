```python
# Definition for singly-linked list.
class ListNode:
     def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def merge_sort(self,nums):
        if len(nums)<2:
            return nums
        mid = len(nums)//2
        left = self.merge_sort(nums[:mid])
        right = self.merge_sort(nums[mid:])
        i = 0
        j = 0
        result = []
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        result += left[i:]
        result += right[j:]
        return result 
    
    def sortList(self, head: ListNode):
        result = []
        tmp = head
        while head:
            result.append(head.val)
            head = head.next
        result = self.merge_sort(result)
        if not result:
            return head
        head = ListNode(result[0])
        result_list = head
        for i in range(1,len(result)):
            result_list.next = ListNode(result[i])
            result_list = result_list.next
        return head
```

```python
class Solution:
    @staticmethod
    def insertionSort(nums): 
        for i in range(1, len(nums)): 
            up = nums[i] 
            j = i - 1
            while j >=0 and nums[j] > up:  
                nums[j + 1] = nums[j] 
                j -= 1
            nums[j + 1] = up      
        return nums
    
    @staticmethod
    def bucketSort(nums): 
        arr = [] 
        slot_num = 10 # 10 means 10 slots, each 
                      # slot's size is 0.1 
        for i in range(slot_num): 
            arr.append([]) 

        # Put array elements in different buckets  
        for j in nums: 
            index_b = int(slot_num * j)  
            arr[index_b].append(j) 

        # Sort individual buckets  
        for i in range(slot_num): 
            arr[i] = insertionSort(arr[i]) 

        # concatenate the result 
        k = 0
        for i in range(slot_num): 
            for j in range(len(arr[i])): 
                nums[k] = arr[i][j] 
                k += 1
        return nums
    
    def sortList(self, head: ListNode):
        result = []
        tmp = head
        while head:
            result.append(head.val)
            head = head.next
        result = bucketSort(result)
        if not result:
            return head
        head = ListNode(result[0])
        result_list = head
        for i in range(1,len(result)):
            result_list.next = ListNode(result[i])
            result_list = result_list.next
        return head
```
