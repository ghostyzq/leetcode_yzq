```python
def solution(N, K):
    if N*K<=0 or type(N) is not int or type(K) is not int :
        return [""]
    result = []
    for p in solution(N - 1, K):
        for l in "abc":
            if p[-1:] != l:
                result += [p + l]
    return result[:K]

```

```python
a = solution(4, 100)
```
