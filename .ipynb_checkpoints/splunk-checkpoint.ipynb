```python
def solution(matrix,point):
    row,column= len(matrix[0]),len(matrix)
    result = []
    #visited = []
    current_position = point
    stack = []
    step = 1
    stack.append([point,'right',step])
    while stack:
        tmp_list = stack.pop()
        x,y,direction,step = tmp_list[0][0],tmp_list[0][1],tmp_list[1],tmp_list[2]
        if direction == 'right':
            step += 2
        if direction == 'right':
            if x + step < column:
                result.append(matrix[x+step][y])
                stack.append([x+step,y],'down',step)
            else:
                stack.append([x+step,y],'down',step)
        if direction == 'down':
            if y + step < row:
                result.append(matrix[x][y+step])
                stack.append([x,y+step],'left',step)
            else:
                stack.append([x,y+step],'left',step)
        if direction == 'left':
            if y - step - 1 >= 0:
                result.append(matrix[x][y-step-1])
                stack.append([x,y-step - 1],'up',step)
            else:
                stack.append([x,y-step - 1],'up',step)
        if direction == 'up':
            if x - step - 1 >= 0:
                result.append(matrix[x-step - 1][y])
                stack.append([x-step - 1,y],'right',step)
            else:
                stack.append([x-step - 1,y],'right',step)
    print(result)
    return
```
